<b>Ejercicio 1:</b><br>
Programa que diga el máximo de 3 números dados

```python
num1 = int(input("Introduce un primer número: "))
num2 = int(input("Introduce un segundo número: "))
num3 = int(input("Introduce un tercer número: "))

if((num1 >= num2) & (num1 >= num3)):
  print(f"El número más alto es: {num1}")
elif((num2 >= num1) & (num2 >= num3)):
  print(f"El número más alto es: {num2}")
else:
  print(f"El número más alto es: {num3}")
```

<b>Ejercicio 3:</b><br>
Programa que determine si un carácter previamente dado es vocal o no

```python
letra = input("Introduce una letra: ")
vocales = ('a','e','i','o','u')

print("¡Es una vocal!") if letra in vocales else print("No es una vocal")
```

<b>Ejercicio 4:</b><br>
Programa que sume los valores de una lista

```python
lista = (1,2,3,4,5)
suma = 0
for num in lista:
  suma += num

print(f"La suma de todos los valores es: {suma}")
```

<b>Ejercicio 5:</b><br>
Programa que diga si una palabra dada es palíndromo

```python
palabra = input("Introduce una palabra: ")
esPalindromo = True
index = 0

while ((index) < len(palabra)):
  if(palabra[(index)] != palabra[-(index+1)]):
    esPalindromo = False
  index += 1

print(f"La palabra '{palabra}' es un palindromo") if esPalindromo else print(f"La palabra '{palabra}' NO es un palindromo")
```

<b>Ejercicio 6:</b><br>
Programa que compare dos listas y encuentre si hay algún valor coincidente


```python
lista1 = (1,2,3,4,5)
lista2 = (1,9)
#lista2 = (6,9)

valoresCoincidentes = []
for x in lista1:
  if (x in lista2):
    valoresCoincidentes.append(x)

print(f"Hay valores coincidentes entre las dos listas: {valoresCoincidentes}") if (len(valoresCoincidentes) >= 1) else print("No hay valores coincidentes entre las dos listas")
```
