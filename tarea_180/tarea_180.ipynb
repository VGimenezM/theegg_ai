10. Crea un array que represente una tabla de ajedrez. Negras 1 y blancas 0.

```python
lineaPar = np.array([0,1,0,1,0,1,0,1], dtype = int)
lineaInpar = np.array([1,0,1,0,1,0,1,0], dtype = int)

tablero = np.zeros((8,8), dtype=int)

for i in range(8):
  if (i % 2 == 0):
    tablero[i] = lineaPar
  else:
    tablero[i] = lineaInpar

tablero
```
